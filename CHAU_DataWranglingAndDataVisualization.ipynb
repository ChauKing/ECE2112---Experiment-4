```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
enggeceboardexams = pd.read_excel('board2.xlsx')
enggeceboardexams
```

```python
Instru = enggeceboardexams.loc[(enggeceboardexams ['Track'] == 'Instrumentation') 
                             & (enggeceboardexams ['Hometown'] == 'Luzon') 
                             & (enggeceboardexams ['Electronics'] > 70), ['Name', 'GEAS', 'Electronics']]
Instru
```

```python
enggeceboardexams['Average'] = (enggeceboardexams['Math'] + enggeceboardexams['Electronics'] 
                              + enggeceboardexams['GEAS'] + enggeceboardexams['Communication']) / 4

Mindy = enggeceboardexams.loc[(enggeceboardexams['Hometown']=='Mindanao') 
                            & (enggeceboardexams['Gender']=='Female') 
                            & (enggeceboardexams['Average']>=55), ('Name','Track','Electronics','Average')]
Mindy
```

```python
plt.figure(figsize=(15, 5))
plt.bar(enggeceboardexams['Track'], enggeceboardexams['Average'])
plt.bar(enggeceboardexams['Gender'], enggeceboardexams['Average'])
plt.bar(enggeceboardexams['Hometown'], enggeceboardexams['Average'])

plt.title('Average Grades by Track, Gender, and Hometown')
plt.xlabel('classification')
plt.ylabel('Average Grade')
```
